```python
import time
from typing import TypedDict

import numpy as np
import faiss
from ddgs import DDGS
from bs4 import BeautifulSoup
from playwright.async_api import async_playwright
from langchain_text_splitters import RecursiveCharacterTextSplitter
from sentence_transformers import SentenceTransformer
from groq import Groq, InternalServerError, RateLimitError
from langgraph.graph import StateGraph, START, END

client = Groq(api_key="")
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")
MAX_SEARCHES = 5


def generate(prompt, retries=5, base_delay=2):
    for attempt in range(retries):
        try:
            response = client.chat.completions.create(
                model="llama-3.3-70b-versatile",
                messages=[{"role": "user", "content": prompt}],
            )
            return response.choices[0].message.content
        except (InternalServerError, RateLimitError):
            if attempt == retries - 1:
                raise
            time.sleep(base_delay * (2 ** attempt))


class State(TypedDict):
    query: str
    urls: list
    html_pages: list
    documents: list
    chunks: list
    embeddings: list
    faiss_index: object
    retrieved_chunks: list
    answer: str
    search_count: int
    enough_info: bool


def search_node(state):
    with DDGS() as ddgs:
        results = list(ddgs.text(state["query"], max_results=10))
    state["urls"] = [r.get("href") or r.get("url") for r in results if r.get("href") or r.get("url")]
    return state


async def playwright_node(state):
    html_pages = []
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()
        for url in state["urls"]:
            try:
                await page.goto(url, wait_until="domcontentloaded", timeout=15000)
                html_pages.append({"url": url, "html": await page.content()})
            except Exception:
                continue
        await browser.close()
    state["html_pages"] = html_pages
    return state


def parse_node(state):
    documents = []
    for page in state["html_pages"]:
        soup = BeautifulSoup(page["html"], "html.parser")
        documents.append({"url": page["url"], "content": soup.get_text(separator=" ", strip=True)})
    state["documents"] = documents
    return state


def chunk_node(state):
    splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
    chunks = []
    for doc in state["documents"]:
        chunks.extend(splitter.split_text(doc["content"]))
    state["chunks"] = chunks
    return state


def embedding_node(state):
    state["embeddings"] = embedding_model.encode(state["chunks"])
    return state


def faiss_node(state):
    vectors = np.array(state["embeddings"]).astype("float32")
    index = faiss.IndexFlatL2(vectors.shape[1])
    index.add(vectors)
    state["faiss_index"] = index
    return state


def retrieve_node(state):
    query_vector = np.array(embedding_model.encode([state["query"]])).astype("float32")
    _, I = state["faiss_index"].search(query_vector, 5)
    state["retrieved_chunks"] = [state["chunks"][idx] for idx in I[0]]
    return state


def evaluate_node(state):
    if state["search_count"] >= MAX_SEARCHES:
        state["enough_info"] = True
        return state

    context = "\n".join(state["retrieved_chunks"])
    prompt = f"""Question:
{state['query']}

Context:
{context}

Can the question be answered confidently from this context?
Reply only YES or NO."""
    state["enough_info"] = "YES" in generate(prompt).strip().upper()
    return state


def route_node(state):
    return "answer" if state["enough_info"] else "research"


def research_node(state):
    prompt = f"""Create a better search query.

Question:
{state['query']}"""
    state["query"] = generate(prompt)
    state["search_count"] += 1
    return state


def answer_node(state):
    context = "\n\n".join(state["retrieved_chunks"])
    prompt = f"""You are a helpful research assistant.

Answer the question ONLY using the provided context.
If the answer is not present in the context, say:
"I could not find sufficient information in the retrieved documents."

Context:
{context}

Question:
{state['query']}"""
    state["answer"] = generate(prompt)
    return state


def final_node(state):
    print(state["query"])
    print(state["answer"])
    return state


graph = StateGraph(State)

graph.add_node("search", search_node)
graph.add_node("playwright", playwright_node)
graph.add_node("parse", parse_node)
graph.add_node("chunk", chunk_node)
graph.add_node("embed", embedding_node)
graph.add_node("faiss", faiss_node)
graph.add_node("retrieve", retrieve_node)
graph.add_node("evaluate", evaluate_node)
graph.add_node("research", research_node)
graph.add_node("answer", answer_node)
graph.add_node("final", final_node)

graph.add_edge(START, "search")
graph.add_edge("search", "playwright")
graph.add_edge("playwright", "parse")
graph.add_edge("parse", "chunk")
graph.add_edge("chunk", "embed")
graph.add_edge("embed", "faiss")
graph.add_edge("faiss", "retrieve")
graph.add_edge("retrieve", "evaluate")
graph.add_conditional_edges("evaluate", route_node, {"answer": "answer", "research": "research"})
graph.add_edge("research", "search")
graph.add_edge("answer", "final")
graph.add_edge("final", END)

app = graph.compile()

await app.ainvoke({"query": "How much wheat is produced in Madhya Pradesh?", "search_count": 0})
```
